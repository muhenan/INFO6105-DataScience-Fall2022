## Use bayesian to find the best parameters of the rainfall dataset

```python
from pymc3 import Model, Exponential, Gamma, fit, sample, Normal
import pymc3 as pm
RANDOM_SEED = 20090425
```

We use normal distributions to model parameter alpha and beta

```python
with Model() as rainfall_model:
    ualpha = Normal('ualpha', 10, sd=10)
    ubeta = Normal('ubeta', 5, sd=10)
```

Now go ahead and model our dataset in pymc3:

```python
with rainfall_model:
    april_like = Gamma('april_like', alpha=ualpha, beta=ubeta, observed=precip["Apr"])
```

Fit the model

```python
with rainfall_model:
    rainfall_trace = pm.sample()
```

```python
pm.traceplot(rainfall_trace)
```

Now plot all the distributions of parameters

```python
from pymc3 import plot_posterior
plot_posterior(rainfall_trace[100:], 
                var_names=['ualpha', 'ubeta'],
                color='#87ceeb');
```

## Add MLE and MOM

### MLE

```python
from scipy.special import psi, polygamma

# Note that log_mean and mean_log are parameters of the dlgamma function
dlgamma = lambda m, log_mean, mean_log: np.log(m) - psi(m) - log_mean + mean_log
dl2gamma = lambda m, *args: 1./m - polygamma(1, m)
```

```python
log_mean = precip.mean().apply(np.log)
mean_log = precip.apply(np.log).mean()
```

```python
from scipy.optimize import newton
```

```python
alpha_mle = newton(dlgamma, 2, dl2gamma, args=(log_mean[3], mean_log[3]))
alpha_mle
```

```python
beta_mle = alpha_mle/precip.mean()[3]
beta_mle
```

For bayesian simulation and MLE

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

from scipy.stats.distributions import gamma

precip.Apr.hist(density=True, bins=10, grid=False)
x = np.linspace(0, 12)
plt.plot(x, gamma.pdf(x, np.mean(rainfall_trace['ualpha']), np.mean(rainfall_trace['ubeta'])), 'r-', label='Bayesian')
plt.plot(x, gamma.pdf(x, alpha_mle, beta_mle), 'g-', label='MLE')
plt.legend()
```

### MOM

```python
precip_mean = precip.mean()
precip_mean

precip_var = precip.var()
precip_var

alpha_mom = precip_mean ** 2 / precip_var
beta_mom = precip_var / precip_mean
```

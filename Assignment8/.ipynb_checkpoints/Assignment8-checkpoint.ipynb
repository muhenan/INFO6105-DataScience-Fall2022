```python
import pandas as pd
```

```python
precip = pd.read_table("nashville_precip.txt", index_col=0, na_values='NA', delim_whitespace=True)
precip.head()
```

```python
precip.fillna(value={'Apr': precip.Oct.mean()}, inplace=True)
```

```python
precip.Apr
```

```python
from pymc3 import Model, Exponential, Gamma, fit, sample, Normal
import pymc3 as pm
RANDOM_SEED = 20090425
```

```python
with Model() as rainfall_model:
    ualpha = Normal('ualpha', 10, sd=10)
    ubeta = Normal('ubeta', 5, sd=10)
```

```python
with rainfall_model:
    april_like = Gamma('april_like', alpha=ualpha, beta=ubeta, observed=precip["Apr"])
```

```python
with rainfall_model:
    #rainfall_trace = fit(random_seed=RANDOM_SEED).sample(2000)
    rainfall_trace = pm.sample()
```

```python
pm.traceplot(rainfall_trace)
```

```python
from pymc3 import plot_posterior
plot_posterior(rainfall_trace[100:], 
                var_names=['ualpha', 'ubeta'],
                color='#87ceeb');
```

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

from scipy.stats.distributions import gamma

precip.Apr.hist(density=True, bins=10, grid=False)
x = np.linspace(0, 12)
plt.plot(x, gamma.pdf(x, np.mean(rainfall_trace['ualpha']), np.mean(rainfall_trace['ubeta'])), 'r-')
```
